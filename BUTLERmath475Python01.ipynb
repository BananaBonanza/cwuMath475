#### 1) Getting Started: First plots and imports

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t=np.linspace(0,3*np.pi, 100) #100 evenly distributed points betweeen 0 and 3pi
x=np.cos(t)
y=np.sin(t)
z=np.zeros(t.shape)

plt.figure() #a figure is a top-level container for plotting elements

plt.plot(t,x, color='k', linestyle='--', label='$\cos(x)$') #dollar signs are used for LaTeX math mode
plt.plot(t,y, color='k', label='$\sin(x)$') #label = is for legend
plt.plot(t,z, color='k', linestyle=':') #plot y=0 in dots
plt.title('Sine and Cosine', fontsize=20) #A title for the graph
plt.xlabel('$x$', fontsize=16)
plt.ylabel('$y$', fontsize=16)
plt.legend(fontsize=16)
plt.xlim(0,3*np.pi) #xlim sets the size of the graph window
plt.show() #display the current figure
plt.close() #close the current figure in order to save memory
```

#### 2) Overlaid Graphs: The process of putting graphs on other graphs. (Without explicit code to type)

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t=np.linspace(0,4*np.pi, 150) #150 evenly distributed points between 0 and 4pi
x=t*np.cos(t)
y=t*np.sin(t)
z=np.zeros(t.shape)

plt.figure() #top level command for plotting elements

plt.plot(t,x, color='k', linestyle='--', label='$t\cos(t)$') #label cosine function
plt.plot(t,y, color='k', label='$t\sin(t)$') #no linestyle, since this plot will be a solid line
plt.plot(t,z, color='k', linestyle=':') #x-axis
plt.title('$t\sin(t)$ and $t\cos(t)$ from 0 to $4\pi$')
plt.xlabel('$t$', fontsize=16)
plt.ylabel('$t\sin(t)$ and $t\cos(t)$')
plt.legend(loc=1,fontsize=16)
plt.xlim=(0,4*np.pi) #Window set in the x direction
plt.ylim=(-4*np.pi,4*np.pi) #Window set in the y direction
plt.show()
plt.close()
```

#### 3) Stacked Plots: Putting graphs on seperate graphs rather than overlaying them.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

t=np.linspace(0,4*np.pi, 150) #150 evenly distributed points between 0 and 4pi
x=t*np.cos(t)
y=t*np.sin(t)
z=np.zeros(t.shape)

plt.figure() #begin plot

#Plotting the first graph with label and limits
plt.subplot(2,1,1)
plt.plot(t,x, color='k', linestyle='--', label='$t\cos(t)$')
plt.title('$t\cos(t)$')
plt.xlabel('$t$')
plt.ylabel('$t\cos(t)')
plt.axis([0, 4*np.pi, -4*np.pi, 4*np.pi])

#Changing the distances between the two graphs
fig=plt.figure()
fig.subplots_adjust(hspace=0.1)

#Plotting the second graph with labels and limits
plt.subplot(2,1,2)
plt.plot(t,y, color='k', label='$t\sin(t)$')
plt.title('$t\sin(t)$')
plt.xlabel('$t$')
plt.ylabel('$t\sin(t)$ and $t\cos(t)$')
plt.axis([0, 4*np.pi, -4*np.pi, 4*np.pi])

plt.show() #display the graph
plt.close() #close the graph to save memory
```

#### 4) Splish-splash: Replacing numbers with strings

```python
outstr = ''
for i in range(150): # i = 0, 1, 2, 3, ..., 149
    j = i+1
    if j%5 == 0:
        outstr = outstr+ "splish, "
    elif j%7 == 0:
        outstr = outstr+ "splash, "
    else:
        partstr = '{0:d}, '.format(j)
        outstr = outstr+partstr
    if j%15 == 0:
        outstr = outstr + '\n'
print (outstr) #Syntax error without the parenthesis. I know this isn't verbatim anymore, but it wouldn't run without it.
```

Multiples of both 5 and 7 are being replaced with a 'splish'. This is because the code first sees that the number is divisible by 5 and sets the string to add a 'splish'. It then never gets to the option of adding a 'splash' and instead goes on to the next number.

```python
outstr = ''
for i in range(150): # i = 0, 1, 2, 3, ..., 149
    j = i+1
    #Setting end condition of 150. Since this is on the top of the if/elseif chain, it will always check this first.
    if j == 150:
        outstr = outstr+ "splish."
    elif j%5 == 0:
        outstr = outstr+ "splish, "
    elif j%7 == 0:
        outstr = outstr+ "splash, "
    else:
        partstr = '{0:d}, '.format(j)
        outstr = outstr+partstr
    if j%15 == 0:
        outstr = outstr + '\n'
print (outstr) #Syntax error without the parenthesis.
```

#### 5) The Bim-Bam challenge: Making our own loop of if/else interchanges.

```python
outstr = ''
for i in range(100): # Setting the for loop to run 100 times
    j = i+1
    
    #Setting the end condition of 100.
    if j == 100:
        outstr = outstr+ "BimBam."
        
    #Setting BimBam for multiples of both 3 and 5
    elif j%15 == 0:
        outstr = outstr+ "BimBam, "
        
    #Setting Bim for multiples of 3
    elif j%3 == 0:
        outstr = outstr+ "Bim, "
        
    #Setting Bam for multiples of 5
    elif j%5 == 0:
        outstr = outstr+ "Bam, "
        
    #Setting all of the other numbers
    else:
        partstr = '{0:d}, '.format(j)
        outstr = outstr+partstr
        
print (outstr) #print the string of numbers/bim/bam/bimbam
```
